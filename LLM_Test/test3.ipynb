```python
def to_seconds(time_val):
    if pd.isnull(time_val): return None
    
    # 숫자를 문자열로 변환하고 길이에 따라 처리
    time_val = str(int(time_val)).zfill(6)  # 최소 6자리로 채움, 예: 011101 -> 00시11분01초
    
    if len(time_val) == 4:  # 4자리 (HHMM)
        hh, mm = int(time_val[:2]), int(time_val[2:])
        ss = 0  # 초는 0으로 설정
    elif len(time_val) == 6:  # 6자리 (HHMMSS)
        hh, mm, ss = int(time_val[:2]), int(time_val[2:4]), int(time_val[4:])
    
    return hh * 3600 + mm * 60 + ss

# 시간 계산
filtered_data['탑승초'] = filtered_data['card_brd_tm'].apply(to_seconds)
filtered_data['하차초'] = filtered_data['alit_deal_tm'].apply(to_seconds)

# 날짜 차이 계산 (같은 날이면 0, 다음 날이면 +1일 고려)
filtered_data['탑승일'] = pd.to_datetime(filtered_data['brd_deal_dt'], format='%Y%m%d', errors='coerce')
filtered_data['하차일'] = pd.to_datetime(filtered_data['alit_deal_dt'], format='%Y%m%d', errors='coerce')

# 전체 초 계산
filtered_data['탑승총초'] = filtered_data['탑승일'].astype('int64') // 10**9 + filtered_data['탑승초']
filtered_data['하차총초'] = filtered_data['하차일'].astype('int64') // 10**9 + filtered_data['하차초']

# 소요 시간 계산
filtered_data['소요시간(초)'] = filtered_data['하차총초'] - filtered_data['탑승총초']
```

```python
from collections import defaultdict
from statistics import mean
# (출발, 도착)이 다를 경우만 처리
temp = defaultdict(list)
for start, end, time in data:
    if start != end:
        temp[(start, end)].append(time)

# 평균 소요시간 계산
avg_data = []
for (start, end), times in temp.items():
    avg_time = round(mean(times))
    avg_data.append((start, end, avg_time))

# 출발 정류장 기준 그룹핑 및 시간순 정렬
grouped = defaultdict(list)
for start, end, avg_time in avg_data:
    grouped[start].append((end, avg_time))

for start in grouped:
    grouped[start].sort(key=lambda x: x[1])

# 출력
for start in sorted(grouped.keys()):
    print(f"[{start}]")
    for end, avg_time in grouped[start]:
        print(f"  └→ {end} : {avg_time}초")
```

```python
test_list_bus = []
test_list_time = []

for start in sorted(grouped.keys()):
    test_list_1 = []
    test_list_2 = []
    test_list_1.append(start)
    idx = 0
    for end, avg_time in grouped[start]:
        test_list_1.append(end)
        if idx == 0:
            test_list_2.append(avg_time)
            last_time = avg_time
        else:
            test_list_2.append(avg_time - last_time)
            last_time = avg_time
        idx += 1
    test_list_bus.append(test_list_1)
    test_list_time.append(test_list_2)
```

```python
# 유추하는 알고리즘
from collections import deque

# 1. 승차/하차 정류장이 포함된 정류장 쌍 추출
station_pairs = list(filtered_data[['승차버스정류장명', '하차버스정류장명']].dropna().itertuples(index=False, name=None))

# 2. 정류장 순서 리스트 초기화 (처음엔 비어 있음)
station_order = deque()

# 3. 순서 유추 로직
for 승차, 하차 in station_pairs:
    # 둘 다 리스트에 없는 경우 → 승차 먼저, 하차 다음
    if 승차 not in station_order and 하차 not in station_order:
        station_order.append(승차)
        station_order.append(하차)

    # 승차만 있고 하차는 없는 경우 → 승차 뒤에 하차 삽입
    elif 승차 in station_order and 하차 not in station_order:
        idx = station_order.index(승차)
        station_order.insert(idx + 1, 하차)

    # 하차만 있고 승차는 없는 경우 → 하차 앞에 승차 삽입
    elif 하차 in station_order and 승차 not in station_order:
        idx = station_order.index(하차)
        station_order.insert(idx, 승차)

    # 둘 다 있는 경우 → 순서가 잘못됐으면 교정
    elif 승차 in station_order and 하차 in station_order:
        if station_order.index(승차) > station_order.index(하차):
            # 위치가 잘못됐을 경우: 현재 순서 제거 후 다시 삽입
            station_order.remove(하차)
            idx = station_order.index(승차)
            station_order.insert(idx + 1, 하차)

# 4. 결과 출력
print("유추된 정류장 순서:")
for i, station in enumerate(station_order, 1):
    print(f"{i}. {station}")
```

```python
import pandas as pd

# 시간 문자열을 초 단위로 변환하는 함수
def to_seconds(time_val):
    if pd.isnull(time_val): return None
    time_val = str(int(time_val)).zfill(6)
    hh, mm, ss = int(time_val[:2]), int(time_val[2:4]), int(time_val[4:])
    return hh * 3600 + mm * 60 + ss

# 시간 변환
test['탑승초'] = test['card_brd_tm'].apply(to_seconds)

# 차량별로 시간 정렬
test = test.sort_values(by=['card_car_cd', '탑승초']).reset_index(drop=True)

# 다음 정류장과 시간 구하기
test['다음정류장'] = test.groupby('card_car_cd')['승차버스정류장명'].shift(-1)
test['다음탑승초'] = test.groupby('card_car_cd')['탑승초'].shift(-1)

# 정류장 간 소요 시간
test['소요시간(초)'] = test['다음탑승초'] - test['탑승초']

# 유효한 구간만 필터링
df_valid = test[
    (test['다음정류장'].notna()) &
    (test['소요시간(초)'] > 0)
]

# 정류장 간 평균 소요 시간 계산
avg_times = (
    df_valid
    .groupby(['승차버스정류장명', '다음정류장'])['소요시간(초)']
    .mean()
    .reset_index()
)

avg_times['평균소요시간(분)'] = (avg_times['소요시간(초)'] / 60).round(2)

# 결과 정리
avg_times = avg_times.rename(columns={
    '승차버스정류장명': '출발',
    '다음정류장': '도착'
})[['출발', '도착', '평균소요시간(분)']]

# 출력
print(avg_times)
```
